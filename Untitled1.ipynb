```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
dataset = pd.read_csv('../datasets/studentscores.csv')
```

```python
# Display the dataset
print(dataset)
```

```python
# Sort the dataset by 'Scores' column in descending order
df = dataset.sort_values("Scores", ascending=False)
print(df)
```

```python
# Display the first 30 rows of the dataset
dataset.head(30)
```

```python
# Extracting features (X) and target variable (Y)
X = dataset.iloc[0:25, :1].values  # Features: 0th to 24th row, 0th column
Y = dataset.iloc[0:25, -1:].values  # Target variable: 0th to 24th row, last column
print("X:", X)
print("Y:", Y)

```

```python
from sklearn.model_selection import train_test_split

# Split the data, using 25% as the test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
print(X_train, X_test)
print(Y_train, Y_test)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize and train the Linear Regression model using the training data
regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

```python
# Predict using the test set
Y_pred = regressor.predict(X_test)
print(Y_pred)
print(Y_test)
```

```python
# Scatter plot of the training data
plt.scatter(X_train, Y_train, color='red')
# Plotting the regression line
plt.plot(X_train, regressor.predict(X_train), 'bo-')
plt.show()
```

```python
# Scatter plot of the test data
plt.scatter(X_test, Y_test, color='red')
# Plotting the predicted values
plt.plot(X_test, Y_pred, 'bo-')
plt.show()
```
